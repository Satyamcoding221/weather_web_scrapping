```python
for year in range(2009,2019):
    for month in range(1,13):
        ind=[]
        mon_yr = int(str(year)+str(month).rjust(2,'0'))
        stock_url = f'http://www.estesparkweather.net/archive_reports.php?date={mon_yr}'
        #stock_url  = 'http://www.estesparkweather.net/archive_reports.php?date=200901'
        #headers = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.117 Safari/537.36'}
        response = requests.get(stock_url)       #, headers=headers)
        soup = BeautifulSoup(response.text, 'html.parser')
        data_array = soup.find("table").getText()
        result = data_array.split("\n")
        real_data = []
        for i in result:
          if i == '':
            continue
          else:
            real_data.append(i)
        real_data = real_data[1:]
        fi = df.columns
        t=[]
        for i in range(0, len(fi)):
          tem = real_data[i].split(' ')
          #print(tem)
          temp1 = fi[i].split(' ')
          b= len(temp1)
          t.append(tem[b-1])
        df.loc[len(df.index)] = t
```
